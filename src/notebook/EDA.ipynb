```python
import os
import pandas as pd
import geopandas as gpd
```

```python
# read bk17 geojson data into geopd
file_path = '/Users/luohy/Documents/Projects/Treefolio-indexing/qgis/BK17_treefolio.geojson'
gdf = gpd.read_file(file_path)

gdf.head()
```

```python
# list all the fields here

print(gdf.info())
```

# Community Board Tree Report

```python
# Community Board Tree Canopy Cover -- the sum of canopy area
sum_canopy_area = gdf['CanopyArea'].sum()
print(f"Total Canopy Area: {sum_canopy_area:.2f}")

# Total Tree Canopy Volume -- the sum of canopy volume
sum_canopy_volume = gdf['CanopyVolume'].sum()
print(f"Total Canopy Volume: {sum_canopy_volume:.2f}")

# Total Number of Trees
sum_trees = len(gdf)
print(f"Total Number of Trees: {sum_trees}")

# Number of Street Trees
sum_street_trees = (gdf['hasTreeCensusID'] == True).sum()
per_street_trees = format(sum_street_trees / sum_trees, ".2%")
print(f"Total Number of Street Trees: {sum_street_trees}, {per_street_trees}")

# Number of Non-street Trees
sum_nonstreet_trees = (gdf['hasTreeCensusID'] == False).sum()
per_nonstreet_trees = format(sum_nonstreet_trees / sum_trees, ".2%")
print(f"Total Number of Non-street Trees: {sum_nonstreet_trees}, {per_nonstreet_trees}")
```

# Average Individual Tree


```python
# Average Canopy Area
avg_canopy_area = gdf['CanopyArea'].mean()
print(f"Average Canopy Area: {avg_canopy_area:.2f}")

# Average Canopy Volume
avg_canopy_volume = gdf['CanopyVolume'].mean()
print(f"Average Canopy Volume: {avg_canopy_volume:.2f}")

# Average Top of Canopy Height
avg_top_of_canopy_height = gdf['TopofCanopyHeight'].mean()
print(f"Average Top of Canopy Height: {avg_top_of_canopy_height:.2f}")

# Average HighTempHours_Avg_ShadedArea_Ground
gdf['HighTempHours_Avg_ShadedArea_Ground'] = pd.to_numeric(gdf['HighTempHours_Avg_ShadedArea_Ground'], errors='coerce')
avg_HighTempHours_Avg_ShadedArea_Ground = gdf['HighTempHours_Avg_ShadedArea_Ground'].mean()
print(f"Average Canopy Shaded Area on Ground(11am-3pm, HighTempHours): {avg_HighTempHours_Avg_ShadedArea_Ground:.2f}")

# Average HighTempHours_Avg_PerCanopy_StreetShade
gdf['HighTempHours_Avg_Perc_Canopy_StreetShade'] = pd.to_numeric(gdf['HighTempHours_Avg_Perc_Canopy_StreetShade'], errors='coerce')
avg_HighTempHours_Avg_PerCanopy_StreetShade = gdf['HighTempHours_Avg_Perc_Canopy_StreetShade'].mean()
print(f"Average Percent of Canopy Shading Streets(11am-3pm, HighTempHours): {avg_HighTempHours_Avg_PerCanopy_StreetShade:.2f}%")

# Average HighTempHours_Avg_PercCanopy_InShade
gdf['HighTempHours_Avg_Perc_Canopy_InShade'] = pd.to_numeric(gdf['HighTempHours_Avg_Perc_Canopy_InShade'], errors='coerce')
avg_HighTempHours_Avg_Perc_Canopy_InShade = gdf['HighTempHours_Avg_Perc_Canopy_InShade'].mean()
print(f"Average Percent of Canopy Shaded by Buildings (11am-3pm, HighTempHours): {avg_HighTempHours_Avg_Perc_Canopy_InShade:.2f}%")
```

# Community Board Shading Data

```python
# Total shaded area (relative noon)
gdf['RelNoon_ShadedArea'] = pd.to_numeric(gdf['RelNoon_ShadedArea'], errors='coerce')
gdf['RelNoon_ShadedArea_Ground'] = pd.to_numeric(gdf['RelNoon_ShadedArea_Ground'], errors='coerce')
sum_RelNoon_ShadedArea = gdf['RelNoon_ShadedArea'].sum()
sum_RelNoon_ShadedArea_Ground = gdf['RelNoon_ShadedArea_Ground'].sum()
print(f"Total Shaded Area (relative noon): {sum_RelNoon_ShadedArea:.2f}")
print(f"Total Shaded Area on Ground(relative noon): {sum_RelNoon_ShadedArea_Ground:.2f}")

# Average Percent of Shaded Area on Streets (relative noon)
gdf['RelNoon_Perc_Canopy_StreetShade'] = pd.to_numeric(gdf['RelNoon_Perc_Canopy_StreetShade'], errors='coerce')
avg_RelNoon_Perc_Canopy_StreetShade = gdf['RelNoon_Perc_Canopy_StreetShade'].mean()
print(f"Average Percent of Canopy Shading Streets (relative noon): {avg_RelNoon_Perc_Canopy_StreetShade:.2f}%" )

# Ave. Shaded Area Ground (11am-3pm)
gdf['HighTempHours_Avg_ShadedArea_Ground'] = pd.to_numeric(gdf['HighTempHours_Avg_ShadedArea_Ground'], errors='coerce')
avg_HighTempHours_Avg_ShadedArea_Ground = gdf['HighTempHours_Avg_ShadedArea_Ground'].mean()
print(f"Average Shaded Area on Ground (11am-3pm, HighTempHours): {avg_HighTempHours_Avg_ShadedArea_Ground:.2f}")

# Total HighTempHours_Avg_ShadesArea_Ground
sum_HighTempHours_Avg_ShadedArea_Ground = gdf['HighTempHours_Avg_ShadedArea_Ground'].sum()
print(f"Total Shaded Area on Ground (11am-3pm, HighTempHours): {sum_HighTempHours_Avg_ShadedArea_Ground:.2f}")
```

# Shape Area

```python
import geopandas as gpd

def calculate_shapefile_area(filepath):
    # Read the shapefile
    gdf = gpd.read_file(filepath)
    
    # Calculate the total area of all polygons in the shapefile
    total_area = gdf['geometry'].area.sum()
    
    return total_area

# Path to your shapefile (replace 'path_to_your_shapefile.shp' with your actual file path)
filepath = '../../Boundaries/BK17_boundary/BK17_boundary.shp'

# Calculate and print the total area
total_area = calculate_shapefile_area(filepath)
print(f"The total area of the shapefile is: {total_area} square units")
```
